### Scatter plot of velocity, with Gaussian noise

Originally created for a PHY 230 lab question.

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Mon Nov 24, 2025

@author: veazeyj
Assisted by ChatGPT
"""

import matplotlib.pyplot as plt
import numpy as np

# Sample data
t = np.linspace(0, 1, 21)

g = 9.8
v0 = 0
y0 = 2

# Select or program function you want to plot. Comment out the rest.
v_1 = v0 + g*t  # Free fall
v_2 = 7.5*t + 3*t**2  # subtle quadratic
# v = -np.abs(-g * t + 5)

# Add Gaussian noise -- Comment out if you do not want noise.
noise_std_1 = 0.2  # Standard deviation of the noise
v_1 = v_1 + np.random.normal(0, noise_std_1, t.shape)

noise_std_2 = 0.1  # Standard deviation of the noise
v_2 = v_2 + np.random.normal(0, noise_std_2, t.shape)



# Styling variables
marker_size = 30
marker_line_width = 1
font_size = 14
tick_font_size = 16  # Increased tick label font size

# Create a figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot v vs t 
ax1.scatter(t, v_1, label='$v(t)$', facecolors='none', edgecolors='b', s=marker_size, linewidth=marker_line_width)
ax1.set_xlabel('Time (s)', fontsize=font_size)
ax1.set_ylabel('Velocity (m/s)', fontsize=font_size)
#ax1.grid(True)

# Set axis limits
ax1.set_ylim(-0.05, 10.05)
ax1.set_xlim(-0.05, 1.05)

# Tick Mark Parameters
# ax1.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax1.tick_params(axis='x', which='major', left=True, labelleft=True, labelsize=tick_font_size)
ax1.tick_params(axis='y', which='both', left=True, labelleft=True, labelsize=tick_font_size)


# Plot v vs t(optional second plot)
ax2.scatter(t, v_2, label='$v(t)$', facecolors='none', edgecolors='b', s=marker_size, linewidth=marker_line_width)
ax2.set_xlabel('Time (s)', fontsize=font_size)
ax2.set_ylabel('Velocity (m/s)', fontsize=font_size)
#ax1.grid(True)

# Set axis limits
ax2.set_ylim(-0.05, 10.05)
ax2.set_xlim(-0.05, 1.05)

# Tick Mark Parameters
# ax2.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax2.tick_params(axis='x', which='major', left=True, labelleft=True, labelsize=tick_font_size)
ax2.tick_params(axis='y', which='both', left=True, labelleft=True, labelsize=tick_font_size)

# Add legend
#ax1.legend(fontsize=font_size)

plt.tight_layout()
plt.show()
```
