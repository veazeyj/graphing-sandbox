### $v_x$ and $v_y$ scatter plots, semi-quantitative, with Gaussian noise

Originally created for a PHY 230 lab question.

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 29 16:24:47 2024

@author: veazeyj
Assisted by ChatGPT
"""

import matplotlib.pyplot as plt
import numpy as np

font_size=36
marker_size=200
marker_line_width=4

# Sample data
t = np.linspace(0, 1, 21)

# Select the v_x and v_y pairs you want to output; comment out rest
#---these functions for 220 projectile motion---
#v_x = 2 + 0 * t
#v_y = -9.8 * t + 5
#v_x = 4*t
#v_y = -np.abs(-9.8 * t + 5)
v_x = -4*t
#v_y = -20 * (t - 0.5)**2
#---these functions for 220 freefall---
#v_y = 50*(t-0.5)**3
#v_y = 9.8 * t - 5
v_y = -np.abs(-9.8 * t + 5)

# Add Gaussian noise -- Comment out if you do not want noise.
noise_std = 0.1  # Standard deviation of the noise
v_x = v_x + np.random.normal(0, noise_std, t.shape)
v_y = v_y + np.random.normal(0, noise_std, t.shape)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot v_x vs t with noise on the left
ax1.scatter(t, v_x, label='$v_x$', facecolors='none', edgecolors='k',s=marker_size,linewidth=marker_line_width)
ax1.set_xlabel('Time',fontsize=font_size)
ax1.set_ylabel('$v_x$', fontsize=font_size)
#ax1.set_title('$v_x$ vs $t$')
ax1.set_ylim(-5, 5)
#ax1.set_xlim(0, 20)
ax1.grid(True)
# Set custom tick labels, keeping only the zero tick
ax1.set_xticks([0])  # Only show tick at 0 on x-axis
ax1.set_xticklabels(['0'])
ax1.set_yticks([0])  # Only show tick at 0 on y-axis
ax1.set_yticklabels(['0'])

# Remove borders (spines)
#ax1.spines['top'].set_visible(False)
#ax1.spines['right'].set_visible(False)
#ax1.spines['bottom'].set_visible(False)
#ax1.spines['left'].set_visible(False)

# Plot v_y vs t with noise on the right
ax2.scatter(t, v_y, label='$v_y$', facecolors='none', edgecolors='blue',s=marker_size,linewidth=marker_line_width)
ax2.set_xlabel('Time',fontsize=font_size)
ax2.set_ylabel('$v_y$',fontsize=font_size)
#ax2.set_title('$v_y$ vs $t$')
ax2.set_ylim(-5, 5)
#ax2.set_xlim(0, 20)
ax2.grid(True)
# Set custom tick labels, keeping only the zero tick
ax2.set_xticks([0])  # Only show tick at 0 on x-axis
ax2.set_xticklabels(['0'])
ax2.set_yticks([0])  # Only show tick at 0 on y-axis
ax2.set_yticklabels(['0'])

# Remove borders (spines)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
