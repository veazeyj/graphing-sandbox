### Comparing data points with "theory" value.

This plot was created for the PHY 220 freefall experiment but may be adapted.

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Aug  1 12:45:30 2024

@author: veazeyj
Assisted by ChatGPT
"""

import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator

# Sample data for trials and accelerations
trials = [1, 2, 3, 4, 5]  # Trial numbers
accelerations = [9.2, 9.8, 9.6, 11.0, 9.9]  # Acceleration values in m/s

# Corresponding errors for each data point
errors = [0.08, 0.7, 0.3, 0.1, 0.1]  # Errors in m/s

# Custom sizes
marker_size = 10  # Size of the markers
font_size = 22  # Font size for labels and title

# Create the plot with error bars
plt.figure(figsize=(7, 5))

# Define custom labels for x-axis
labels = ['Velocity 1', 'Velocity Analysis', 'Velocity 3', 'Position Analysis', 'Velocity 5']

plt.errorbar(trials, accelerations, yerr=errors, fmt='none', ecolor='none', 
             capsize=5, capthick=1, markerfacecolor='none', markeredgecolor='b', markersize=marker_size)


# Define which ticks to keep
visible_ticks = [2, 4]  # Only keep ticks for 'Velocity 2' and 'Velocity 4'

# Set x-ticks to custom labels
#plt.xticks(ticks=trials, labels=labels)

# Set all tick positions and labels, then hide unwanted ones
ax = plt.gca()
ax.set_xticks(trials)  # Set all trial numbers as ticks
ax.set_xticklabels(labels)  # Set all labels

# Hide ticks and labels for trials 1, 3, and 5
for tick in ax.get_xticks():
    if tick not in visible_ticks:
        # Hide ticks
        ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Set axis labels and title with custom font sizes
plt.xlabel('', fontsize=font_size)
plt.ylabel('Acceleration (m/s)', fontsize=font_size)
plt.title('Comparing data with accepted value', fontsize=font_size)

# Set tick parameters with custom font size
plt.tick_params(axis='both', which='major', labelsize=font_size)

# Set the y-axis limits
plt.ylim(8.5, 11.5)

# Set x-axis to display only whole numbers
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

# Add minor ticks for the y-axis only
plt.gca().yaxis.set_minor_locator(AutoMinorLocator())

# Enable grid for major ticks and minor ticks only on the y-axis
plt.grid(True, which='major', linestyle='-', linewidth=0.75)
plt.grid(True, which='minor', axis='y', linestyle='--', linewidth=0.5)

# Add a horizontal dashed line at y=9.804 with a label
y_value = 9.804
plt.axhline(y=y_value, color='k', linestyle='--', linewidth=2)  # Increased linewidth

rot = 0

plt.text(5.3, y_value, '$g = $9.804 m/s$^2$', color='k', fontsize=font_size-6, 
         ha='left', va='center')  # Added label with appropriate positioning
#plt.text(1.7, y_value-1.5, 'velocity\nanalysis', color='k', fontsize=font_size-6, 
#         ha='left', va='center', rotation=rot)  # Added label with appropriate positioning
#plt.text(3.7, y_value-1.5, 'position\nanalysis', color='k', fontsize=font_size-6, 
#         ha='left', va='center', rotation=rot)  # Added label with appropriate positioning

plt.annotate('velocity\nanalysis', xy=(2, 8.5), xytext=(2, 7.7),
             arrowprops=dict(facecolor='black', shrink=0.05, width=2, headwidth=9),
             fontsize=font_size, ha='center')

plt.annotate('position\nanalysis', xy=(4, 8.5), xytext=(4, 7.7),
             arrowprops=dict(facecolor='black', shrink=0.05, width=2, headwidth=9),
             fontsize=font_size, ha='center')

# Display the plot
plt.show()
```
