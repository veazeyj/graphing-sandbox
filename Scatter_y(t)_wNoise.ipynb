### Scatter plot of position, with Gaussian noise

Originally created for a PHY 230 lab question.

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Mon Nov 24, 2025

@author: veazeyj
Assisted by ChatGPT
"""

import matplotlib.pyplot as plt
import numpy as np

font_size=36
marker_size=200
marker_line_width=4

# Sample data
t = np.linspace(0, 1, 21)

g = 9.8
v0 = 5
y0 = 0

# Select or program function you want to plot. Comment out the rest.
y = -0.5*g*t**2 + v0*t + y0  # Free fall
# y = -np.abs(-g * t + 5)

# Add Gaussian noise -- Comment out if you do not want noise.
noise_std = 0.02  # Standard deviation of the noise
y = y + np.random.normal(0, noise_std, t.shape)



# Styling variables
marker_size = 70
marker_line_width = 1.5
font_size = 14
tick_font_size = 16  # Increased tick label font size

# Create a figure
fig, ax1 = plt.subplots(figsize=(5, 5))

# Plot y vs t with noise
ax1.scatter(t, y, label='$y(t)$', facecolors='none', edgecolors='b', s=marker_size, linewidth=marker_line_width)
ax1.set_xlabel('Time (s)', fontsize=font_size)
ax1.set_ylabel('Position (m)', fontsize=font_size)
#ax1.grid(True)

# Set axis limits
# ax1.set_ylim(0, 2)
# ax1.set_xlim(0, 1)

# Tick Mark Parameters
# ax1.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax1.tick_params(axis='x', which='major', left=True, labelleft=True, labelsize=tick_font_size)
ax1.tick_params(axis='y', which='both', left=True, labelleft=True, labelsize=tick_font_size)


# Add legend
#ax1.legend(fontsize=font_size)

plt.tight_layout()
plt.show()
```
