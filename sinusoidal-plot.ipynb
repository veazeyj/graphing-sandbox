```python

import matplotlib.pyplot as plt
import numpy as np  
from matplotlib.ticker import AutoMinorLocator

# Use Times New Roman globally
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'

label_font = 22
tick_font = 22

### set initial parameters for sinusoidal function ###
x0 = 0
A = 60
T = 12
phi = np.pi/3
t_max = 12
t_min = 0

# Use these for blank axes
t_blank = []
x_blank = []

########Use the following if you'd like to add a plot: the first two arrays are for functions, the second two are for discrete points.########
t = np.linspace(t_min, t_max, 100)
x = x0+A*np.cos(2*np.pi*t/T+phi)
#t = [0, 0.2, 0.3, 0.3, 0.5, 0.6]
#x = [-3, -3, -3, -6, -6, -6]

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))
#ax.plot(t_blank, x_blank, color='blue', linewidth=4)
line1, = ax.plot(t, x, color='green', linewidth=2)
line1.set_visible(True)  # Hide the line by making argument False. Reveal with True

x_min=np.min(t)
x_max=np.max(t)
y_min=np.min(x)
y_max=np.max(x)
# Axis limits
#ax.set_xlim(x_min, x_max)
#ax.set_ylim(y_min, y_max)

# Grid on major + minor ticks
ax.grid(True, which='major', linestyle='--', linewidth=0.8, alpha=0.6)
##ax.grid(True, which='minor', linestyle=':', linewidth=0.6, alpha=0.4)


# Move spines so axes cross at (0,0)
ax.spines['left'].set_position('zero')   # y-axis at t = 0
ax.spines['bottom'].set_position('zero') # x-axis at x = 0

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add axis labels at the ends
ax.text(x_max+x_max/10, 0.0, r'$\it{t}\ (\mathrm{s})$', fontsize=label_font, fontname='Times New Roman')
#ax.text(-0.4, 4, r'$\it{x}\ (\mathrm{m})$', fontsize=label_font)
ax.text(-1*x_max/10, y_max+y_max/5, r'$\it{v_x}\ (\mathrm{cm/s})$', fontsize=label_font)
#ax.text(-0.4, 4, r'$\it{a}\ (\mathrm{m/s^2})$', fontsize=label_font)
#ax.text(-0.4, 540, r'$\it{F_{floor\rightarrow ball}}\ (\mathrm{N})$', fontsize=label_font)

# TICK MARK OPTIONS #
ax.tick_params(axis='both', which='major', labelsize=tick_font, labelbottom=True, labelleft=True)  # major ticks

ax.set_xticks([3, 6, 9, 12])
ax.set_yticks([-60, -30, 0, 30, 60])
# Alternate Ticks (comment out two rows above)
#ax.set_xticks([0,1,2,3,4])
#ax.set_yticks([-10, -5, 0, 5, 10])
#custom_labels_x = ['', '', '$T$', '', '$2T$']
#custom_labels_y = ['$-A$', '', '0', '', '$A$']
#ax.set_xticklabels(custom_labels_x)
#ax.set_yticklabels(custom_labels_y)

ax.xaxis.set_minor_locator(AutoMinorLocator(2))
#ax.yaxis.set_minor_locator(AutoMinorLocator())

plt.show()
```
