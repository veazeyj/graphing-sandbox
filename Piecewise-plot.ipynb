```python

import matplotlib.pyplot as plt

label_font = 14
tick_font = 14

# Data points
t = [0, 2, 4, 6]
x = [0, -4, -4, 6]
#t = []
#x = []

# Create the plot
fig, ax = plt.subplots(figsize=(6, 8))
ax.plot(t, x, color='blue', linewidth=2)

# Axis limits
ax.set_xlim(0, 6.5)
ax.set_ylim(-5, 7)

# Ticks
ax.set_xticks([0, 2, 4, 6])
ax.set_yticks([-4, -2, 0, 2, 4, 6])
# Alternate Ticks (comment out two rows above)
#ax.set_xticks([])
#ax.set_yticks([])

# Move spines so axes cross at (0,0)
ax.spines['left'].set_position('zero')   # y-axis at t = 0
ax.spines['bottom'].set_position('zero') # x-axis at x = 0

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add axis labels at the ends
ax.text(6.6, 0.0, r'$ t \ (\mathrm{s})$', fontsize=label_font)
ax.text(0.3, 6.5, r'$x \ (\mathrm{m})$', fontsize=label_font)

ax.tick_params(axis='both', which='major', labelsize=tick_font)  # major ticks
#ax.tick_params(axis='both', which='minor', labelsize=tick_font-2)  # minor ticks


plt.show()
```
