```python

import matplotlib.pyplot as plt
import numpy as np  

# Use Times New Roman globally
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'

label_font = 14
tick_font = 14

# Data points
#t = [0, 2, 3, 5, 6]
#x = [0, 10, 10, 0, 0]
t = [0, 5]
x = [0, 5]

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(t, x, color='blue', marker='o', markersize=6, linewidth=2)

# Manual Axis Limits
#ax.set_xlim(0, 6.5)
#ax.set_ylim(-5, 7)

# Ticks
ax.set_xticks([0, 5, 10, 15, 20, 25])
ax.set_yticks([0, 5, 10, 15])
# Alternate Ticks (comment out two rows above)
#ax.set_xticks([])
#ax.set_yticks([])

# Move spines so axes cross at (0,0)
#ax.spines['left'].set_position('zero')   # y-axis at t = 0
#ax.spines['bottom'].set_position('zero') # x-axis at x = 0

# Hide top and right spines
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)

# Add axis labels at the ends
ax.text(1.02, 0.0, r'$ x \ (\mathrm{m})$',
        transform=ax.transAxes,
        fontsize=label_font,
        va='center', ha='left')
ax.text(0.0, 1.02, r'$y \ (\mathrm{m})$',
        transform=ax.transAxes,
         fontsize=label_font,
         va='bottom', ha='center')

ax.grid(True, which='major', linestyle='--', linewidth=0.6, alpha=0.6)

ax.tick_params(axis='both', which='major', labelsize=tick_font)  # major ticks
#ax.tick_params(axis='both', which='minor', labelsize=tick_font-2)  # minor ticks

# Label the data points
ax.annotate(r'$t=0$', (t[0], x[0]),
            textcoords="offset points",
            xytext=(12, 3),
            fontsize=12)

ax.annotate(r'$t=1$', (t[1], x[1]),
            textcoords="offset points",
            xytext=(12, 3),
            fontsize=12)

plt.show()
```
