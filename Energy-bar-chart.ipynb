### Energy bar chart template.

This plot was created for a PHY 220 pre-lab but may be adapted.

```python

import matplotlib.pyplot as plt
import numpy as np

fontlabel = 11

# Data
categories = ['$KE_{trans}$', '$KE_{rot}$', '$PE$']
initial_energies = ([0, 0, 3],[1, 1, 3],[0, 0, 3],[0, 0, 3])  # Initial energies
final_energies = ([2, 1, 0],[3, 2, 0],[3, 0, 0],[0, 3, 0])    # Final energies

# Bar positions
x_initial = np.arange(len(categories))
x_final = x_initial + len(categories) + 1

# Create the figure and subplots
fig, axes = plt.subplots(1, 4, figsize=(11, 3.5), sharey=True)
subplot_labels = ['A', 'B', 'C', 'D']

for i, ax in enumerate(axes):
    # Plot the bars
    bars_initial = ax.bar(x_initial, initial_energies[i], width=0.8, label='Initial', color='black')
    bars_final = ax.bar(x_final, final_energies[i], width=0.8, label='Final', color='grey')

    if i == 0:
        ax.set_ylabel('Energy Value')
        ax.legend()

    ax.text(len(categories) / 2 - 0.7, -1.5, 'Initial \n system energy', ha='center', va='center', fontsize=fontlabel)
    ax.text(len(categories) + len(categories) / 2 + 0.7, -1.5, 'Final \n system energy', ha='center', va='center', fontsize=fontlabel)

    ax.set_xticks(np.concatenate([x_initial, x_final]))
    ax.set_xticklabels(categories * 2, rotation=45, fontsize=fontlabel)
    
    # Add horizontal dashed lines at each integer y value
    y_ticks = np.arange(0, max(max(initial_energies[i]), max(final_energies[i])) + 1, 1)
    for y in ([1,2,3,4]):
      ax.axhline(y=y, linestyle='--', color='grey', alpha=0.5, linewidth=0.5)
      
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([]) # keep the ticks but remove the labels

    ax.text(0.05, 0.95, subplot_labels[i], transform=ax.transAxes, fontsize=12, fontweight='normal', va='top')

fig.tight_layout()
plt.show()
```
