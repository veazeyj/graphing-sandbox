```python

import matplotlib.pyplot as plt

# Use Times New Roman globally
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'

label_font = 14
tick_font = 14

# Data points
t = []
x = []
#t = []
#x = []

# Create the plot
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(t, x, color='blue', linewidth=4)

# Axis limits
ax.set_xlim(0, 5.5)
ax.set_ylim(-6.5, 6.5)

# Ticks
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_yticks([-6, -4, -2, 0, 2, 4, 6])
# Alternate Ticks (comment out two rows above)
#ax.set_xticks([])
#ax.set_yticks([])

# Move spines so axes cross at (0,0)
ax.spines['left'].set_position('zero')   # y-axis at t = 0
ax.spines['bottom'].set_position('zero') # x-axis at x = 0

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add axis labels at the ends
ax.text(5.6, -0.2, r'$\it{t}\ (\mathrm{s})$', fontsize=label_font, fontname='Times New Roman')
ax.text(-0.4, 7, r'$\it{v}\ (\mathrm{m/s})$', fontsize=label_font)

ax.tick_params(axis='both', which='major', labelsize=tick_font, labelbottom=False, labelleft=False)  # major ticks
#ax.tick_params(axis='both', which='minor', labelsize=tick_font-2)  # minor ticks


plt.show()
```
