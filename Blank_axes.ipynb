```python

import matplotlib.pyplot as plt
import numpy as np

# Use Times New Roman globally
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'

label_font = 14
tick_font = 14

# Data points
t = [1, 2, 3, 4, 5, 5, 6]
y = [-3, -2, -1, 0, 1, 2, 3]
#t = []
#y = []

# Create the plot
fig, ax = plt.subplots(figsize=(4, 4))
#ax.plot(t, y, color='blue', linewidth=2)
line1, = ax.plot(t, y, color='blue', linewidth=4, linestyle='-')
line1.set_visible(False)  # Hide the line

t_min=np.min(t)
t_max=np.max(t)
y_min=np.min(y)
y_max=np.max(y)

# Axis limits
ax.set_xlim(min(t_min, 0), max(t_max, 0))
ax.set_ylim(min(y_min, 0), max(y_max, 0))

# Ticks
#ax.set_xticks([0, 1, 2, 3, 4, 5])
#ax.set_yticks([-2,2])
# Alternate Ticks (comment out two rows above)
#ax.set_xticks([])
#ax.set_yticks([])

# To hide tick labels, change "labelbottom" and/or "labelleft" to False.
ax.tick_params(axis='both', which='major', labelsize=tick_font, labelbottom=False, labelleft=False)  # major ticks

# Move spines so axes cross at (0,0)
ax.spines['left'].set_position('zero')   # y-axis at t = 0
ax.spines['bottom'].set_position('zero') # x-axis at x = 0

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add axis labels at the ends
ax.text(t_max+t_max/20, 0.0, r'$\it{t}\ (\mathrm{s})$', fontsize=label_font, fontname='Times New Roman')
ax.text(-1*t_max/10, y_max+y_max/10, r'$\it{y}\ (\mathrm{m})$', fontsize=label_font)
### These for wave motion graphs ###
#ax.text(3.5, 2.5, r'At $\it{x=0}\ \mathrm{cm}$', fontsize=label_font)
#ax.text(2, 1, r'$\mathrm{100~cm/s}\ \leftarrow$', fontsize=label_font)

plt.show()
```
